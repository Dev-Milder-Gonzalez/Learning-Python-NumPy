# Creaci√≥n de arrays en NumPyüß©

````python
import numpy as np
print(f'Versi√≥n de la librer√≠a usada: {np.__version__}')
````

## 1. La funci√≥n m√°s importante de NumPy: ```np.array()``` üß±
```np.array(object, dtype=None, copy=True, order='K', subok=False, ndmin=0)```

 - __Object:__ Es el objeto iterable de entrada que queremos convertir en un array de Numpy 
 - __dtype:__ Indica el tipo de dato del array. Si se pone el tipo de datos, Numpy lo infiere autom√°ticamente.
 - __copy:__ Define su NumPy copia los datos o solo usa la referencia.
 - __Order:__ Define como se ordenan los datos en memoria.
 - __subok:__ Indica si NumPy debe permitir subclases de ```ndarray```.
 - __ndmin:__ Define en n√∫mero m√≠nimo de dimensiones que tendr√° el array.

## 1.1. Creaci√≥n desde listas o tuplasüü¢

````python
# Lista 1D -> Array 2D
lista = range(20, 30)
array_1d = np.array(lista)
print(f'Array NumPy:\n{array_1d}')
print('-' * 40)
# Atributos que nos permiten ver las caracter√≠sticas del array de NumPy
print(f'Dimesiones: {array_1d.shape}\nTipo de datos del array: {array_1d.dtype}')
````

````python
# Lista anidada -> array 2D
datos_ventas= [
    [101, "2025-01-01", 1250.5],
    [102, "2025-01-02", 1890.0],
    [103, "2025-01-03", 1670.4]
]
ventas_array = np.array(datos_ventas)
print(f'Array 2D de NumPy:\n{ventas_array}')
print('-' * 40)
print(f'Dimesiones: {ventas_array.shape}\nTipo de dato: {ventas_array.dtype}')
````

__NOTAüí°:__ Si mezclamos tipos de datos diferentes (```int```+ ```srt```), NumPy fuerza todo a ```object(``` o ```String```.

## 1.2. Forzar el tipo de datos (```dtype```)‚úçÔ∏è

````python
# ‚úÖ Separar columnas o forzar tipos
ids = np.array([101, 103, 104], dtype= np.int32)
fechas = np.array(["2025-01-01", "2025-01-02", "2025-01-03"], dtype='U10')  # Unicode 10 chars
importes = np.array([1250.5, 1890.0, 1670.75], dtype=np.float64)
print(f'Ids:\n{ids}')
print('-' * 45)
print(f'Fechas:\n{fechas}')
print('-' * 45)
print(f'Importes:\n{importes}')  
````

````python
# KPIs
print(f'Ingreso total: {importes.sum()}')
print(f'Media diaria: {importes.mean()}')
````

````python
# Dtypes para fechas
fechas_np = np.array(fechas, dtype = "datetime64[D]")
print(f'Array NumPy de fechas:\n{fechas_np}')
````

````python
#Forzar dimesiones (ndadmin)-√∫ltil para broadcasting futuro
vector = np.array(range(1,5), ndmin= 2) # -> shape (1, 4) en vez de (4,)
matriz = np.array([range(1,3), range(3, 5)], ndmin= 3)
print(f'Dimesiones del vector: {vector.shape}')
print(f'Dimesiones de la matriz: {matriz.shape}')
````

## 1.3. Arrays prerellenados‚úÖ
Firmas de los m√©todos a usar:
```
np.zeros(shape, dtype=float, order='C')
np.ones(shape, dtype=None, order='C')
np.full(shape, fill_value, dtype=None, order='C')
```

### 1.3.1. Inicializar acumuladores o contadores: ```zeros()```

````python
#Array para inicializar contadores o aculumuladores
acumulador_mensual = np.zeros(12, dtype= np.float64) #12 meses
print(f'Acumulador mensual:\n{acumulador_mensual}')
````

````python
#Matriz para resultados (por ejemplo 100 clientes x 5 m√©tricas)
matriz_resultados = np.zeros((80 , 5), dtype = np.float64)
print(f'Matriz de resultados (80 clientes por 5 m√©tricas):\n{matriz_resultados}')
````

### 1.3.2. Multiplicaci√≥n rapida o inicializar probabilidades: ```ones()```‚ú®

````python
# Array de unos para multiplicar
precios_unitarios = np.ones(50, dtype= np.float64) 
print(f'Array para los precios unitarios:\n{precios_unitarios}')
````

````python
# Matriz de probabilidades iniciales
probabilidades = np.ones((10, 4), dtype= np.float64)
print(f'Matriz para probabilidades iniciales:\n{probabilidades}')
````

### 1.3.3. Arreglo con un valor constante definido: ```full()```‚úçÔ∏è

````python
# Valores missing (ausencia de datos real) o sentinelas (Indicar el estado o condici√≥n de un elememnto dentro de un conjuto de datos)
missing = np.full(100, fill_value= np.nan)
missing_int_missing = np.full(100, fill_value= -999)

print(f'Primer array:\n{missing}')
print('=' * 80)
print(f'Segundo array:\n{missing_int_missing}')
````

````python
#Inicializar categorias
categorias = np.full(50, fill_value= "Sin_categoria", dtype = "U20")
print(f'Array para categorias:\n{categorias}')
````

````python
# Precio con valor por defecto
precios = np.full((500, 3), fill_value= 0.0)
# 500 productos para tres tiendas
print(f'matriz de precios:\n{precios}')
````

##  1.4. Creaci√≥n de arreglos sin inicializar valores: M√©todo ```np.empty()```üïµÔ∏è
Crear un array sin incializar sus valores, usarlo cuando vayamos a __sobreescribir__ __todos__ los valores __inmediatamente__.
```
numpy.empty(shape, dtype=float, order='C', like=None)
```


````python
temp = np.empty(1_000, dtype= np.float64)
for i in range(1_000):
    temp[i] = i + 1

print(f'Array "temp":\n{temp}')
````
